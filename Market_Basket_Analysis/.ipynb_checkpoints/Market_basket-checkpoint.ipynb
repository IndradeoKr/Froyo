```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("BreadBasket_DMS.csv")
```

```python
to_drop = df.loc[(df['Item']=="NONE")].index

df.drop(to_drop, axis = 0, inplace = True)
df["Item"].unique()
```

```python
df.shape
```

```python
data = df.groupby('Transaction', as_index=False)['Item'].count()
data = data.pivot_table(index='Item', values='Transaction', aggfunc='count')
data
```

```python
ax = sns.barplot(x=data.index, y='Transaction' , data=data, palette=sns.color_palette())
plt.xlabel('Number of Items', size = 12)
plt.ylabel('Count of Transactions', size = 12)
plt.title('Number of Items per Transaction', size = 15)
plt.show()
```

```python
value_counts = df.groupby(['Item']).size().reset_index(name='frequency')
value_counts_sorted = value_counts.sort_values(by='frequency', ascending=False).head(5)
print(value_counts_sorted)
```

```python
bars = value_counts_sorted["Item"]
height = value_counts_sorted["frequency"]

plt.pie(height, labels=bars, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(np.linspace(0, 1, len(bars))))
plt.title("Top 5 sold items listed")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

# def apri(data):
encoding = df.groupby(['Transaction','Item'])['Item'].count().unstack().reset_index().fillna(0).set_index('Transaction').astype(int)
encoding = encoding.applymap(encode_units)
frequent_itemsets = apriori(encoding, min_support=0.01, use_colnames=True)
rules_1 = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
output = rules_1.sort_values(by=['confidence'], ascending=False)
encoding.head(10)
```
